```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
# Holdout validation
cars_numeric.shape
```

```python
def knn_train_test(train_col, target_col, df):
    split = round(len(df)/2) #finds midpoint of rows of df
    np.random.seed(0) 
    
    shuffled_index = np.random.permutation(df.index) #generates series of shuffled indexes
    shuffled_df = df.reindex(shuffled_index) #reindexes df using the shuffled index series 
    
    test = shuffled_df.iloc[:split] #splitting df in half by rows
    train = shuffled_df.iloc[split:]
    
    k_rmse_values = {}
    
    for k in range(1,11,2):
        
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train[[train_col]], train[[target_col]])
        predictions = knn.predict(test[[train_col]])
    
        rmse = round(mean_squared_error(test[target_col], predictions)**(1/2),1)
        k_rmse_values[k] = rmse
    
    return k_rmse_values
```

```python
columns = cars_numeric.columns #columns to train and test with
rmses = {}
columns = columns.drop(['price','engine_type']) #price is the target, engine_type is categorical data
columns
```

```python
#creates dict of rmses for each category with k values [1,3,5,7,9]
for column in columns:
    value = knn_train_test(column, 'price', cars_numeric)
    rmses[column] = value
```

```python
def knn_train_test_multivariate(train_columns, target_col, df):#can pass a list of columns to train_col
    split = round(len(df)/2) 
    np.random.seed(0) 
    
    shuffled_index = np.random.permutation(df.index) 
    shuffled_df = df.reindex(shuffled_index) 
    
    test = shuffled_df.iloc[:split] 
    train = shuffled_df.iloc[split:]
    
    #k_rmse_values = {}
    
    #for k in range(1,11,2):
        
    knn = KNeighborsRegressor(n_neighbors = 5)
    knn.fit(train[train_columns], train[[target_col]])
    predictions = knn.predict(test[train_columns])
    
    rmse = round(mean_squared_error(test[target_col], predictions)**(1/2),1)
    #k_rmse_values[k] = rmse
    
    return rmse
```

```python
#multivariate testing with categories that yielded lowest RMSE values
multi = (
    ['horsepower', 'width'],
    ['horsepower', 'width', 'curb_weight'],
    ['horsepower', 'width', 'curb_weight', 'city_mpg'],
    ['horsepower', 'width', 'curb_weight', 'city_mpg', 'highway_mpg'],
    ['horsepower', 'width', 'curb_weight', 'city_mpg', 'highway_mpg', 'length']
)

multivariate = {}
for i, columns in enumerate(multi): #can't pass lists in multi as a dict key so passing index
    multivariate[i] = knn_train_test_multivariate(columns,'price',cars_numeric)
```

```python
#now should look for best k value given this column set
from sklearn.metrics import mean_absolute_error
best_columns = ['horsepower', 'width', 'curb_weight', 'city_mpg', 'highway_mpg']
def knn_multivariate_multiple_k(train_col, target_col, df):
    split = round(len(df)/2)
    np.random.seed(0) 
    
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index) 
    
    test = shuffled_df.iloc[:split] 
    train = shuffled_df.iloc[split:]
    
    k_rmse_values = {}
    k_mse_values = {}
    k_mae_values = {}
    for k in range(1,21):
        
        knn = KNeighborsRegressor(n_neighbors = k)
        knn.fit(train[train_col], train[[target_col]])
        predictions = knn.predict(test[train_col])
        
        mae = round(mean_absolute_error(predictions, test['price']),1)
        mse = round(mean_squared_error(test[target_col], predictions),1)
        rmse = round(mean_squared_error(test[target_col], predictions)**(1/2),1)
        k_rmse_values[k] = rmse
        k_mse_values[k] = mse
        k_mae_values[k] = mae
        
    
    return k_rmse_values, k_mse_values, k_mae_values
```

```python
best_k = list(knn_multivariate_multiple_k(best_columns,'price',cars_numeric))
best_k
#dict of dicts for k_rmse_values, k_mse_values and k_mae_values
```
